```python
# 멜론 클로링 순위 100 곡 추출
# pip install selenium 수행

from selenium import webdriver          # 브라우저 연결
from bs4 import BeautifulSoup as bs     # 인터넷으로 긁어온 데이터 정제
import pandas as pd                     # 데이터 분석

driver = webdriver.Chrome('C:/Users/EZEN/Documents/PythonClass/chromedriver.exe')    # 드라이버 제어
url = 'https://www.melon.com/chart/'
driver.get(url)
html = driver.page_source
soup = bs(html)

song_data = []
rank = 1

songs = soup.select('tbody > tr')

for song in songs:
    title = song.select('div.rank01 > span > a')[0].text
    singer = song.select('div.rank02 > a')[0].text
    song_data.append(['Melon',rank, title, singer])
    rank += 1

# 이중 리스트로 만든 100곡을 데이터프레임으로 전환하기
df = pd.DataFrame(song_data, columns = ['서비스', '순위', '노래제목', '가수명'])
df

# 데이터프레임을 엑셀로 젖장하기
df.to_excel('Melon.xlsx', index = False)
```
